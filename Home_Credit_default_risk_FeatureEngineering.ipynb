<h2 style = "color : Black"><center>Business Problem </center> </h2>

The Lending Organizations are faced with the tough task of analyzing the risk associated with each client. Therefore, it is important to identify the risky behaviors of clients and make educated decisions. The two most critical questions in the banking industry are 
- (i) How risky is the borrower? 

- (ii) Given the borrower’s risk, should we lend him/her?

Many people struggle to get loans due to insufficient or non-existent credit histories. And, unfortunately, this population is often taken advantage of by untrustworthy lenders. Here we face issue of CREDIT RISK.

Default Credit risk is the chance that companies or individuals will be unable to make the required payments on their debt obligations. In other words, credit default risk is the probability that if you lend money, there is a chance that they won’t be able to give the money back on time. Lenders and investors are exposed to default risk in virtually all forms of credit extensions. To mitigate the impact of default risk, lenders often impose charges that correspond to the debtor’s level of default risk. A higher level of risk leads to a higher required return. 

Many people struggle to get loans due to insufficient or non-existent credit histories. And, unfortunately, this population is often taken advantage of by untrustworthy lenders. Doing so will ensure that clients capable of repayment are not rejected and that loans are given with a principal, maturity, and repayment calendar that will empower their clients to be successful.
So, Lending Organizations are approached usually by TWO types of customers:

- 1.	People who do not particularly have a prior credit history, for example students’ loans, small businessmen, etc. who need credits, be it for studies, or for setting up some sort of businesses. Without adequate credit history, the lending organizations find it difficult to lend credits to such people, as these loans could be associated with high risks. In these kinds of situations, some lending organizations even tend to exploit the borrowers by asking for too high of an interest rate.
- 2.	People, who do have prior credit history, which could be with the same organization or some other organizations. However, going through that historical data could be very time consuming and redundant. This would scale up even further as the number of applicants increases. For such cases, if there could be a way through which the lending organization could predict or estimate the borrower’s repayment capability, the process could be streamlined and be made effective for both the lender and the borrower. It could save resources both in terms of humans and time.

All client's previous credits provided by other financial institutions that were reported to Credit Bureau (for clients who have a loan in our sample).For every loan in our sample, there are as many rows as number of credits the client had in Credit Bureau before the application date.

**bureau_balance.csv:**

- Monthly balances of previous credits in Credit Bureau.
- This table has one row for each month of history of every previous credit reported to Credit Bureau – i.e the table has (#loans in sample * # of relative previous credits * # of months where we have some history observable for the previous credits) rows.

**POS_CASH_balance.csv:**

- Monthly balance snapshots of previous POS (point of sales) and cash loans that the applicant had with Home Credit.
- This table has one row for each month of history of every previous credit in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample * # of relative previous credits * # of months in which we have some history observable for the previous credits) rows.

**credit_card_balance.csv:**

- Monthly balance snapshots of previous credit cards that the applicant has with Home Credit.
- This table has one row for each month of history of every previous credit in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample * # of relative previous credit cards * # of months where we have some history observable for the previous credit card) rows.

**previous_application.csv:**

- All previous applications for Home Credit loans of clients who have loans in our sample.
- There is one row for each previous application related to loans in our data sample.

**installments_payments.csv:**

- Repayment history for the previously disbursed credits in Home Credit related to the loans in our sample.
- There is a) one row for every payment that was made plus b) one row each for missed payment.
- One row is equivalent to one payment of one instalment OR one instalment corresponding to one payment of one previous Home Credit related to loans in our sample.

```python
## preprocessing the train and test file
### ** There is a catch actually we need to preprocess the train and test data together as we are filling NA and using label 
## encoding ,Once the train data is transfomed ,The label ENcoder wont work well,SO we will pre process one by one
#we will create function for this



def preporcess_application_data(train,test):
    """
    This function take application_train|test .csv and just do some normal preprocessing
    These are just primary preprocessing
    """
    #Step 1. drop the non important columns
    
    #In EDA we have seen some column which were no use for predicting Target Varibale,due to low variance,and multicollineraity
    #So we will drop it
    col_to_drop=['FLAG_MOBIL',
                 #'REGION_RATING_CLIENT_W_CITY','LIVE_REGION_NOT_WORK_REGION',
                 'FLAG_DOCUMENT_2','FLAG_DOCUMENT_4','FLAG_DOCUMENT_10','FLAG_DOCUMENT_12','FLAG_DOCUMENT_20']
    #For building properties,we will just keep the AVG columns and remove median and mode as those are giving same infoprmation
    #For median values of bulinding properties
    avg= [col for col in train.columns if col.split("_")[-1]=='AVG']
    median_cols= [col for col in train.columns if col.split("_")[-1]=='MEDI']
    #for mode values of bulinding properties
    mode_cols= [col for col in train.columns if col.split("_")[-1]=='MODE']
    non_building_mode_cols=list(set([i.split('_')[0] for i in mode_cols]).difference(set([i.split('_')[0] for i in avg])))

    non_building_mode_cols= [elem+"_MODE" for elem in non_building_mode_cols]

    #non_building_mode_cols

    mode_cols = [ elem for elem in mode_cols if elem not in non_building_mode_cols ]

    
    #add these to col_to_drop
    col_to_drop.extend(median_cols)
    col_to_drop.extend(mode_cols)
    #drop these columns
    train= train.drop(col_to_drop,axis=1)
    test= test.drop(col_to_drop,axis=1)
    #step : 2. 
    
    # here we will convert some column and also remove some outlier and make those NAN
    ##converting age from days to years
    train['YEARS_BIRTH'] = (-1 / 365)*train['DAYS_BIRTH']
    test['YEARS_BIRTH'] = (-1 / 365)*test['DAYS_BIRTH']
    #we can see in test data we dont have 'XNA' category in code_gender,So we can remove such rows,also we just had 4 rows in train dat
    train= train[train['CODE_GENDER'] != 'XNA']
    test= test[test['CODE_GENDER'] != 'XNA']
    
    #in DAYS_EMPLOYED we have some outliers as 365243,we need to remove it or better we can make it NAN
    
    train[train['DAYS_EMPLOYED']==365243]['DAYS_EMPLOYED']=np.nan
    test[test['DAYS_EMPLOYED']==365243]['DAYS_EMPLOYED']=np.nan

    #similary we can do same for SOCIAL CIRCLE columns
    train[train['OBS_30_CNT_SOCIAL_CIRCLE']>30]['OBS_30_CNT_SOCIAL_CIRCLE']=np.nan
    train[train['OBS_60_CNT_SOCIAL_CIRCLE']>30]['OBS_30_CNT_SOCIAL_CIRCLE']=np.nan
    test[test['OBS_30_CNT_SOCIAL_CIRCLE']>30]['OBS_30_CNT_SOCIAL_CIRCLE']=np.nan
    test[test['OBS_60_CNT_SOCIAL_CIRCLE']>30]['OBS_30_CNT_SOCIAL_CIRCLE']=np.nan
    
    
    return train,test
```

## Step 2

```python
# we will try to do encoding for categrical data
# Remember some of the numerical columns were seemed to be categoricals,we will incude those as well
from sklearn.preprocessing import LabelEncoder
def categorical_encoding_and_fillna_continous_application(train,test):
    #NAME_CONTRACT_TYPE
    NAME_CONTRACT_TYPE_encode= {'Cash loans': 1, 'Revolving loans': 0}
    train['NAME_CONTRACT_TYPE']= train['NAME_CONTRACT_TYPE'].map(NAME_CONTRACT_TYPE_encode)
    test['NAME_CONTRACT_TYPE']= test['NAME_CONTRACT_TYPE'].map(NAME_CONTRACT_TYPE_encode)
    
    #CODE_GENDER
    train['CODE_GENDER']= train['CODE_GENDER'].map({'F':1,'M':2})
    test['CODE_GENDER']= test['CODE_GENDER'].map({'F':1,'M':2})
    
    #FLAG_OWN_CAR
    train['FLAG_OWN_CAR']= train['FLAG_OWN_CAR'].map({'N':1,'Y':0})
    test['FLAG_OWN_CAR']= test['FLAG_OWN_CAR'].map({'N':1,'Y':0})
    
    #FLAG_OWN_REALTY
    train['FLAG_OWN_REALTY']= train['FLAG_OWN_REALTY'].map({'N':0,'Y':1})
    test['FLAG_OWN_REALTY']= test['FLAG_OWN_REALTY'].map({'N':0,'Y':1})
    
    #NAME_TYPE_SUITE
    train['NAME_TYPE_SUITE'].fillna('Unaccompanied',inplace=True)
    test['NAME_TYPE_SUITE'].fillna('Unaccompanied',inplace=True)

    le_NAME_TYPE_SUITE=LabelEncoder()
    le_NAME_TYPE_SUITE.fit(train['NAME_TYPE_SUITE'])
    train['NAME_TYPE_SUITE']=le_NAME_TYPE_SUITE.transform(train['NAME_TYPE_SUITE'])
    test['NAME_TYPE_SUITE']=le_NAME_TYPE_SUITE.transform(test['NAME_TYPE_SUITE'])
    
    #NAME_INCOME_TYPE
    le_NAME_INCOME_TYPE=LabelEncoder()
    le_NAME_INCOME_TYPE.fit(train['NAME_INCOME_TYPE'])
    train['NAME_INCOME_TYPE']=le_NAME_INCOME_TYPE.transform(train['NAME_INCOME_TYPE'])
    test['NAME_INCOME_TYPE']=le_NAME_INCOME_TYPE.transform(test['NAME_INCOME_TYPE'])
    #NAME_EDUCATION_TYPE
    map_NAME_EDUCATION_TYPE={'Secondary / secondary special': 4,
                             'Higher education': 2,
                             'Incomplete higher': 3,
                             'Lower secondary': 5,
                             'Academic degree': 1}
    
    train['NAME_EDUCATION_TYPE']= train['NAME_EDUCATION_TYPE'].map(map_NAME_EDUCATION_TYPE)
    test['NAME_EDUCATION_TYPE']= test['NAME_EDUCATION_TYPE'].map(map_NAME_EDUCATION_TYPE)

    #NAME_FAMILY_STATUS
    map_NAME_FAMILY_STATUS= {'Married': 5,
                             'Single / not married': 2,
                             'Civil marriage': 3,
                             'Separated': 2,
                             'Widow': 1,
                             'Unknown': 0}
    train['NAME_FAMILY_STATUS']= train['NAME_FAMILY_STATUS'].map(map_NAME_FAMILY_STATUS)
    test['NAME_FAMILY_STATUS']= test['NAME_FAMILY_STATUS'].map(map_NAME_FAMILY_STATUS)
    
    
    #NAME_HOUSING_TYPE
    map_NAME_HOUSING_TYPE={'House / apartment': 5,
                       'With parents': 4,
                       'Municipal apartment': 2,
                       'Rented apartment': 3, 'Office apartment': 1,
                       'Co-op apartment': 1}
    
    train['NAME_HOUSING_TYPE']= train['NAME_HOUSING_TYPE'].map(map_NAME_HOUSING_TYPE)
    test['NAME_HOUSING_TYPE']= test['NAME_HOUSING_TYPE'].map(map_NAME_HOUSING_TYPE)

    #OCCUPATION_TYPE
    map_OCCUPATION_TYPE= {'IT staff': 0,
                          'HR staff': 0,
                          'Realty agents': 0,
                          'Secretaries': 0,
                          'Waiters/barmen staff': 0,
                          'Private service staff': 0,
                          'Low-skill Laborers': 1,
                          'Cleaning staff': 1,
                          'Accountants': 1,
                          'Medicine staff': 2,
                          'Cooking staff': 2,
                          'High skill tech staff': 2,
                          'Security staff': 2,
                          'Managers': 3,
                          'Core staff': 3,
                          'Drivers': 3,
                          'Sales staff': 4,
                          'Laborers': 5}
    
    
    train['OCCUPATION_TYPE']= train['OCCUPATION_TYPE'].map(map_OCCUPATION_TYPE)
    test['OCCUPATION_TYPE']= test['OCCUPATION_TYPE'].map(map_OCCUPATION_TYPE)

    #filling na values 1 as these occupation with no values can belong to category 1
    train['OCCUPATION_TYPE']= train['OCCUPATION_TYPE'].fillna(1)
    test['OCCUPATION_TYPE']= test['OCCUPATION_TYPE'].fillna(1)
    
    #WEEKDAY_APPR_PROCESS_START: we will just drop this column because its no use
    train.drop('WEEKDAY_APPR_PROCESS_START',axis=1,inplace=True)
    test.drop('WEEKDAY_APPR_PROCESS_START',axis=1,inplace=True)
    
    #ORGANIZATION_TYPE,we can catregorize in it in 3 
    #ORGANIZATION_TYPE
    categories= train['ORGANIZATION_TYPE'].value_counts().index.to_list()
    category_defaulter_dict=dict()
    for i in categories:
        count_defaulter_for_category=train[(train['ORGANIZATION_TYPE']==i) & (train['TARGET']==1)].shape[0]
        count_total_defaulter=train[(train['TARGET']==1)].shape[0]
        #count_total=df_train[df_train[]].shape[0]
        percetage_defaulter= float(count_defaulter_for_category*100/count_total_defaulter)
        category_defaulter_dict[i]=percetage_defaulter
        
    map_org_type={k: v for k, v in sorted(category_defaulter_dict.items(), key=lambda item: item[1])}
    
    map_org_type_encode={}
    for i,j in map_org_type.items():
        if j >= 10:
            map_org_type_encode[i]=2
        elif j<10 and j>=1:
            map_org_type_encode[i]=1
        else:
            map_org_type_encode[i]=0   

    train['ORGANIZATION_TYPE']= train['ORGANIZATION_TYPE'].map(map_org_type_encode)
    test['ORGANIZATION_TYPE']= test['ORGANIZATION_TYPE'].map(map_org_type_encode)

#     #FONDKAPREMONT_MODE
    train['FONDKAPREMONT_MODE'].fillna('Not specified',inplace=True)
    test['FONDKAPREMONT_MODE'].fillna('Not specified',inplace=True)
    
    #we can label encode it
    le_FONDKAPREMONT_MODE=LabelEncoder()
    le_FONDKAPREMONT_MODE.fit(train['FONDKAPREMONT_MODE'])
    train['FONDKAPREMONT_MODE']=le_FONDKAPREMONT_MODE.transform(train['FONDKAPREMONT_MODE'])
    test['FONDKAPREMONT_MODE']=le_FONDKAPREMONT_MODE.transform(test['FONDKAPREMONT_MODE'])

    #HOUSETYPE_MODE
    map_HOUSETYPE_MODE={'Not specified':2,'block of flats':1,'specific housing': 0,'terraced house': 0}
    train['HOUSETYPE_MODE']= train['HOUSETYPE_MODE'].map(map_HOUSETYPE_MODE)
    test['HOUSETYPE_MODE']= test['HOUSETYPE_MODE'].map(map_HOUSETYPE_MODE)

    #filling na as 'Not specified':2
    train['HOUSETYPE_MODE']= train['HOUSETYPE_MODE'].fillna(2)
    test['HOUSETYPE_MODE']= test['HOUSETYPE_MODE'].fillna(2)
    
    #WALLSMATERIAL_MODE
    train['WALLSMATERIAL_MODE']=train['WALLSMATERIAL_MODE'].fillna("Not Specified")
    test['WALLSMATERIAL_MODE']=test['WALLSMATERIAL_MODE'].fillna("Not Specified")

    map_wallsmaterial_type={'Monolithic': 0,
                       'Others': 0,
                       'Mixed': 0,
                       'Wooden': 1,
                       'Block': 1,
                       'Panel': 1,
                       'Stone, brick': 1,
                       'Not Specified':2}
    train['WALLSMATERIAL_MODE']=train['WALLSMATERIAL_MODE'].map(map_wallsmaterial_type)
    test['WALLSMATERIAL_MODE']=test['WALLSMATERIAL_MODE'].map(map_wallsmaterial_type)

        
    #EMERGENCYSTATE_MODE
    train['EMERGENCYSTATE_MODE']=train['EMERGENCYSTATE_MODE'].fillna("No")
    train['EMERGENCYSTATE_MODE']=train['EMERGENCYSTATE_MODE'].map({'Yes':1,'No':0})
    
    test['EMERGENCYSTATE_MODE']=test['EMERGENCYSTATE_MODE'].fillna("No")
    test['EMERGENCYSTATE_MODE']=test['EMERGENCYSTATE_MODE'].map({'Yes':1,'No':0})
    #Filling Na Values for continous columns
    #AMT_GOODS_PRICE : fill it with same value of AMT_credit
    train['AMT_GOODS_PRICE']=train['AMT_GOODS_PRICE'].fillna(train['AMT_CREDIT'])
    test['AMT_GOODS_PRICE']=test['AMT_GOODS_PRICE'].fillna(test['AMT_CREDIT'])
    
    #OWN_CAR_AGE,fillna with zero because they dont have car
    train['OWN_CAR_AGE']= train['OWN_CAR_AGE'].fillna(0)
    test['OWN_CAR_AGE']= test['OWN_CAR_AGE'].fillna(0)

    #DAYS_LAST_PHONE_CHANGE
    train['DAYS_LAST_PHONE_CHANGE']=train['DAYS_LAST_PHONE_CHANGE'].fillna(0)
    test['DAYS_LAST_PHONE_CHANGE']=test['DAYS_LAST_PHONE_CHANGE'].fillna(0)
    #CNT_FAM_MEMBERS
    train['CNT_FAM_MEMBERS']=train['CNT_FAM_MEMBERS'].fillna(0)
    test['CNT_FAM_MEMBERS']=test['CNT_FAM_MEMBERS'].fillna(0)
    #AMT_ANNUITY;we suppose that where AMT_ANNUITY is equal to  AMT_CREDIT without considering interest and repayment
    #period as 1 year
    train['AMT_ANNUITY']=train['AMT_ANNUITY'].fillna(train['AMT_CREDIT'])
    test['AMT_ANNUITY']=test['AMT_ANNUITY'].fillna(test['AMT_CREDIT'])

    #fill na for 'EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3' with 0 i.e zero income from thate externel income source
    train['EXT_SOURCE_1']= train['EXT_SOURCE_1'].fillna(0)
    train['EXT_SOURCE_2']= train['EXT_SOURCE_2'].fillna(0)
    train['EXT_SOURCE_3']= train['EXT_SOURCE_3'].fillna(0)
    test['EXT_SOURCE_1']= test['EXT_SOURCE_1'].fillna(0)
    test['EXT_SOURCE_2']= test['EXT_SOURCE_2'].fillna(0)
    test['EXT_SOURCE_3']= test['EXT_SOURCE_3'].fillna(0)
    #TOTALAREA_MODE: filling it with  highest correlation colum LIVINGAREA_AVG
    #train['TOTALAREA_MODE']= np.where(train['TOTALAREA_MODE'].isnull(), train['LIVINGAREA_AVG'], train['TOTALAREA_MODE'])
    
       #For Avg columns
    building_properties_avg= [col for col in train.columns if col.split("_")[-1]=='AVG']
    for i in building_properties_avg:
        mean= train[i].median()
        train[i]= train[i].fillna(mean)
        test[i]= test[i].fillna(mean)
    #for trains
    
    #AMT_REQ_CREDIT_BUREAU_HOUR    13.501806
    train['AMT_REQ_CREDIT_BUREAU_HOUR'] = train['AMT_REQ_CREDIT_BUREAU_HOUR'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_DAY     13.501806
    train['AMT_REQ_CREDIT_BUREAU_DAY'] = train['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_WEEK    13.501806
    train['AMT_REQ_CREDIT_BUREAU_WEEK'] = train['AMT_REQ_CREDIT_BUREAU_WEEK'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_MON     13.501806
    train['AMT_REQ_CREDIT_BUREAU_MON'] = train['AMT_REQ_CREDIT_BUREAU_MON'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_QRT     13.501806
    train['AMT_REQ_CREDIT_BUREAU_QRT'] = train['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_YEAR    13.501806
    train['AMT_REQ_CREDIT_BUREAU_YEAR'] = train['AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(0)
    # OBS_30_CNT_SOCIAL_CIRCLE       0.332025
    train['OBS_30_CNT_SOCIAL_CIRCLE'] = train['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(0)
    # DEF_30_CNT_SOCIAL_CIRCLE       0.332025
    train['DEF_30_CNT_SOCIAL_CIRCLE'] = train['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(0)
    # OBS_60_CNT_SOCIAL_CIRCLE       0.332025
    train['OBS_60_CNT_SOCIAL_CIRCLE'] = train['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(0)
    # DEF_60_CNT_SOCIAL_CIRCLE       0.332025
    train['DEF_60_CNT_SOCIAL_CIRCLE'] = train['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(0)
    train['TOTALAREA_MODE']= np.where(train['TOTALAREA_MODE'].isna(), train['LIVINGAREA_AVG'], train['TOTALAREA_MODE'])
    #for test
    #AMT_REQ_CREDIT_BUREAU_HOUR    13.501806
    test['AMT_REQ_CREDIT_BUREAU_HOUR'] = test['AMT_REQ_CREDIT_BUREAU_HOUR'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_DAY     13.501806
    test['AMT_REQ_CREDIT_BUREAU_DAY'] = test['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_WEEK    13.501806
    test['AMT_REQ_CREDIT_BUREAU_WEEK'] = test['AMT_REQ_CREDIT_BUREAU_WEEK'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_MON     13.501806
    test['AMT_REQ_CREDIT_BUREAU_MON'] = test['AMT_REQ_CREDIT_BUREAU_MON'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_QRT     13.501806
    test['AMT_REQ_CREDIT_BUREAU_QRT'] = test['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(0)
    # AMT_REQ_CREDIT_BUREAU_YEAR    13.501806
    test['AMT_REQ_CREDIT_BUREAU_YEAR'] = test['AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(0)
    # OBS_30_CNT_SOCIAL_CIRCLE       0.332025
    test['OBS_30_CNT_SOCIAL_CIRCLE'] = test['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(0)
    # DEF_30_CNT_SOCIAL_CIRCLE       0.332025
    test['DEF_30_CNT_SOCIAL_CIRCLE'] = test['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(0)
    # OBS_60_CNT_SOCIAL_CIRCLE       0.332025
    test['OBS_60_CNT_SOCIAL_CIRCLE'] = test['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(0)
    # DEF_60_CNT_SOCIAL_CIRCLE       0.332025
    test['DEF_60_CNT_SOCIAL_CIRCLE'] = test['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(0)
    test['TOTALAREA_MODE']= np.where(test['TOTALAREA_MODE'].isna(), test['LIVINGAREA_AVG'], test['TOTALAREA_MODE'])
    
    return train,test  
    
```

```python
def application_data_feature_engineering(df):
    """
    Here we will create some intersting features,These are basically domain knowledge and some random features
    """
    #1. for client EMployment and client Birth#####################################################
    df['PER_DAYS_EMPLOYED']= df['DAYS_EMPLOYED'] /df['DAYS_BIRTH']
    df['DAYS_UNEMPLOYED'] = abs(df['DAYS_BIRTH'])- abs(df['DAYS_EMPLOYED'])
    df['PER_DAYS_EMPLOYED']= df['DAYS_EMPLOYED'] /df['DAYS_BIRTH']
    df['DAYS_UNEMPLOYED'] = abs(df['DAYS_BIRTH'])- abs(df['DAYS_EMPLOYED'])
    df["OWN_CAR_AGE_RATIO"] = df["OWN_CAR_AGE"] / df["DAYS_BIRTH"]
    df["DAYS_ID_PUBLISHED_RATIO"] = df["DAYS_ID_PUBLISH"] / df["DAYS_BIRTH"]
    df["DAYS_REGISTRATION_RATIO"] = df["DAYS_REGISTRATION"] / df["DAYS_BIRTH"]
    df["DAYS_LAST_PHONE_CHANGE_RATIO"] = df["DAYS_LAST_PHONE_CHANGE"] / df["DAYS_BIRTH"]
    #2. clients INCOME#################################################################
    df['CREDIT_INCOME_RATIO'] = df['AMT_CREDIT'] / (df['AMT_INCOME_TOTAL'] + 0.001)
    df['ANNUITY_INCOME_RATIO'] = df['AMT_ANNUITY'] / (df['AMT_INCOME_TOTAL'] + 0.001)
    df['GOODS_INCOME_RATIO'] = df['AMT_GOODS_PRICE'] / (df['AMT_INCOME_TOTAL'] + 0.001)
    df['INCOME_EXT1_RATIO'] = df['AMT_INCOME_TOTAL'] / (df['EXT_SOURCE_1'] + 0.001)
    df['INCOME_EXT2_RATIO'] = df['AMT_INCOME_TOTAL'] / (df['EXT_SOURCE_2'] + 0.001)
    df['INCOME_EXT3_RATIO'] = df['AMT_INCOME_TOTAL'] / (df['EXT_SOURCE_3'] + 0.001)
    df['INCOME_ANNUITY_DIFF'] = df['AMT_INCOME_TOTAL'] - df['AMT_ANNUITY']
    #percentage income of person and the credit amount
    df['INCOME_PER_CAPITA'] = df['AMT_INCOME_TOTAL'] / df['CNT_FAM_MEMBERS']
    ########For AMT_Credit and AMT_ANNUITY####################################################
    #percentage income of person and the credit amount
    df['INCOME_CREDIT_RATIO'] = df['AMT_INCOME_TOTAL'] / df['AMT_CREDIT']
    #Amount paid for previous loan appication every month decided by the number of day employed
    df['ANNUITY_DAYS_EMPLOYED_PERC'] = df['DAYS_EMPLOYED']/ df['AMT_ANNUITY']
    
    df['AMT_CREDIT_DAYS_EMPLOYED_PERC'] = df['DAYS_EMPLOYED']/ df['AMT_CREDIT']
    #Anually paid amount to amount credited
    df['PAYMENT_RATE'] = df['AMT_ANNUITY'] / df['AMT_CREDIT']
    
    df['PAYMENT_RATE_INV'] = df['AMT_CREDIT'] / df['AMT_ANNUITY']

    df['PAY_TOWARDS_LOAN'] = df['AMT_INCOME_TOTAL']-df['AMT_ANNUITY']
    df['CREDIT_EXT1_RATIO'] = df['AMT_CREDIT'] / (df['EXT_SOURCE_1'] + 0.001)
    df['CREDIT_EXT2_RATIO'] = df['AMT_CREDIT'] / (df['EXT_SOURCE_2'] + 0.001)
    df['CREDIT_EXT3_RATIO'] = df['AMT_CREDIT'] / (df['EXT_SOURCE_2'] + 0.001)

    # FOR OWN_CAR_AGE##########################################################################
    df['CAR_EMPLOYED_DIFF'] = df['OWN_CAR_AGE'] - df['DAYS_EMPLOYED']
    df['CAR_EMPLOYED_RATIO'] = df['OWN_CAR_AGE'] / (df['DAYS_EMPLOYED']+0.00001)
    df['CAR_AGE_DIFF'] = abs(df['DAYS_BIRTH']) - abs(df['OWN_CAR_AGE'])
    df['CAR_AGE_RATIO'] = df['OWN_CAR_AGE'] / (abs(df['DAYS_BIRTH']))
    
    #FOR Family members #############################################################################
    df["CNT_ADULTS"] = df["CNT_FAM_MEMBERS"] - df["CNT_CHILDREN"]
    df['CHILDREN_RATIO'] = (df['CNT_CHILDREN']+0.0001) / (df['CNT_FAM_MEMBERS'])
    df['LOG_CNT_CHILDREN']= np.log( df['CNT_CHILDREN'])
    df['LOG_CNT_FAM_MEMBERS']=np.log(df['CNT_FAM_MEMBERS'])
    #FOR FLAG DOCMENTS
    flag_doc_cols =[col for col in df.columns if "FLAG_DOCUMENT_" in col ]
    df['CNT_FLAG_DOCS']= df[flag_doc_cols].sum(axis=1)
    # for "NOT_LIVE" and "NOT WORK",These are the column where Flag=1 if there is a adress missmatch
    address_missmatch_cols= [col for col in df.columns if ("NOT_LIVE" in col) or  ("NOT_WORK" in col) ]
    df['ADDRESS_MISMATCH']= df[flag_doc_cols].sum(axis=1)
    #Even i know the AMT_REQ_CREDIT_BUREAU are no use here but will just take mean of these columns
    AMT_Req_cb_cols= [col for col in df.columns if "AMT_REQ_CREDIT_BUREAU" in col ]
    df['AMT_REQ_CREDIT_BUREAU_MEAN']= df[flag_doc_cols].mean(axis=1)
    df['AMT_REQ_CREDIT_BUREAU_SUM']= df[flag_doc_cols].sum(axis=1)
    df['AMT_ENQ_CREDIT_RATIO'] = df['AMT_REQ_CREDIT_BUREAU_SUM'] / (df['AMT_CREDIT'] + 0.00001)

    #for Phone/Email contant
    df['All_CONTACTS']=((df[['FLAG_EMP_PHONE','FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE','FLAG_PHONE','FLAG_EMAIL']]).sum(axis=1))
    #for days ID ,Registration changed
    df['MAX_DAYS_CHANGED']=((df[['DAYS_ID_PUBLISH','DAYS_REGISTRATION']]).max(axis=1))
    #EXT_SOURCE_COLUMNS
    df['EXT_SOURCE_SUM']=(df[['EXT_SOURCE_1', 'EXT_SOURCE_2','EXT_SOURCE_3']]).sum(axis=1)
    df['EXT_SOURCE_MEAN']=(df[['EXT_SOURCE_1', 'EXT_SOURCE_2','EXT_SOURCE_3']]).mean(axis=1)
    
    df['EXT_SOURCE_MEDIAN']=(df[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']]).median(axis=1)
    
    df['EXT_SOURCE_MIN']=(df[['EXT_SOURCE_1', 'EXT_SOURCE_2','EXT_SOURCE_3']]).min(axis=1)
    
    df['EXT_SOURCE_MAX']=(df[['EXT_SOURCE_1', 'EXT_SOURCE_2','EXT_SOURCE_3']]).max(axis=1)
    #BUILDING PROPERT SCORE scores,we have only considered the AVG ones because mean 
    avg= [col for col in df.columns if col.split("_")[-1]=='AVG']
    df['BUILDING_PROPERTIES_AVG_SUM'] = df[avg].sum(axis=1)
    #client's social surroundings OBSERVED And DEFAULTED
    social_surr_cols= [col for col in df.columns if "CNT_SOCIAL_CIRCLE" in col ]
    df['CNT_SOCIAL_CIRCLE_MEAN']=df[social_surr_cols].sum(axis=1)
    
    #Now we will create some features based on contionous columns and aggregated by categorical columns
    #We wont create any column which have some reference to "TARGET" VARIABLE,because TEST data wont have it
    #We have the categorcial column and some continous column which are actually categorical column ,We will include both
    # we have seen some interaction between differenet categorical variable ,SO we will try to group on some categorical varaible
    interaction_cols_for_aggregation_on= [['OCCUPATION_TYPE','ORGANIZATION_TYPE'],
                                     ['CODE_GENDER','NAME_CONTRACT_TYPE'],
                                     ['FLAG_OWN_CAR','FLAG_OWN_REALTY'],
                                     ['CODE_GENDER','FLAG_OWN_REALTY'],
                                     ['NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','CODE_GENDER'],
                                     ['NAME_FAMILY_STATUS','NAME_HOUSING_TYPE'],
                                     ['FONDKAPREMONT_MODE','NAME_INCOME_TYPE'],
                                     ['HOUSETYPE_MODE','NAME_HOUSING_TYPE','NAME_INCOME_TYPE'],
                                     ['EMERGENCYSTATE_MODE','NAME_INCOME_TYPE'],['REGION_RATING_CLIENT']]
    
    #Contionious column for aggregation,we are not using all we are just using some important one,
    contiouns_cols_for_agrregation_of={'AMT_ANNUITY' : ['mean','max','min'],
                                       'ANNUITY_INCOME_RATIO' : ['mean','max','min'],
                                       'DAYS_UNEMPLOYED' : ['mean','min'],
                                       'AMT_INCOME_TOTAL' : ['mean','max','min'],
                                       'BUILDING_PROPERTIES_AVG_SUM' : ['mean','max','min'],
                                       'EXT_SOURCE_MEAN' : ['mean','max','min'],
                                       'EXT_SOURCE_1' : ['mean','max','min'],
                                       'EXT_SOURCE_2' : ['mean','max','min'],
                                       'EXT_SOURCE_3' : ['mean','max','min']}
    
    for group_col in interaction_cols_for_aggregation_on:
        #grouping on categorical feature interaction
        grouped_data= df.groupby(group_col).agg(contiouns_cols_for_agrregation_of)
        grouped_data.columns= ['_'.join(i).upper() + '_AGG_BY_' + '_'.join(group_col) for i in grouped_data.columns]
        #merging with data
        df= df.join(grouped_data,on=group_col)
    #Some Log and box cox transform
    log_vars = ["AMT_CREDIT", "AMT_INCOME_TOTAL", "AMT_GOODS_PRICE", "AMT_ANNUITY"]
    for i in log_vars:
        df['LOG_'+str(i)]= np.log(abs(df[i])+1)
    
    box_cox_vars= ['DAYS_EMPLOYED','YEARS_BIRTH','DAYS_REGISTRATION','OWN_CAR_AGE','EXT_SOURCE_1','EXT_SOURCE_1','EXT_SOURCE_1']
    from scipy import stats

    for i in box_cox_vars:
        df['BOXCOX_'+str(i)],_= stats.boxcox(abs(df[i])+1)
    
    #df= pd.get_dummies(df)
    df.columns= ["app_"+col.upper() if col not in ["TARGET","SK_ID_CURR"] else col for col in df.columns]
    
    return df
```

```python
def preprocess_agrregate_bureau_balance(df):
    
    # first we will encode the status column,we are trying to encode it with reference to defaulter seen in EDA
    #when we merged it with train data, we have given C=0 because closed status is better one for Target=0
    #rest are in same sequnetial manner

    map_status= { 'C': 0,
                 '0': 1,
                 '1': 2, 
                 '2': 3, 
                 'X': 4, 
                 '3': 5, 
                 '4': 6, 
                 '5': 7}
    df['STATUS']= df['STATUS'].map(map_status)

    # MONTHS_BALANCE, 
    #As for each customer id data is given in month wise and year wise manner ,we will sort the data in ascending order
    # by each id and month,basically the output will be for each id monthwise data
    df= df.sort_values(by=['SK_ID_BUREAU', 'MONTHS_BALANCE'], ascending=[0, 0])
    #now we have sorted data,instead of Taking monthly data,we can take yearly data
    #first as MONTHS_BALANCE is negative we will trun it into postiive
    df['MONTHS_BALANCE']=  -1*df['MONTHS_BALANCE']
    df['YEAR_BALANCE']= df['MONTHS_BALANCE']//12
    #creating  a feature Loan score 
    df['Loan_score']= (df['STATUS']+1)/(df['MONTHS_BALANCE']+1)

    #now we will agreegate the data by SK_ID_BUREAU and get some aggregated values as well
    bureau_balance_aggregate= df.groupby(['SK_ID_BUREAU']).aggregate({'STATUS':['min','max','mean','first','last'],
                                                                     'MONTHS_BALANCE':['min','max','mean','first','last'],
                                                                     'YEAR_BALANCE':['min','max','mean','first','last'],
                                                                     'Loan_score':'sum'})
    #https://stackoverflow.com/questions/47666438/how-to-groupby-aggregate-on-multiple-columns-and-rename-the-multi-index-in-pandas
    bureau_balance_aggregate.columns= bureau_balance_aggregate.columns.map('_'.join)

    #additionally we 
    #Now our idea is we will take latest data for one year and agreegare it 
    bureau_balance_aggregate_year_0= df[df['YEAR_BALANCE']==0].groupby(['SK_ID_BUREAU']).aggregate({'STATUS':['min','max','mean','first','last'],
                                                                     'MONTHS_BALANCE':['min','max','mean','first','last','sum'],
                                                                     'YEAR_BALANCE':['min','max','mean','first','last','sum'],
                                                                     'Loan_score':'sum'})
    #https://stackoverflow.com/questions/47666438/how-to-groupby-aggregate-on-multiple-columns-and-rename-the-multi-index-in-pandas
    bureau_balance_aggregate_year_0.columns= bureau_balance_aggregate_year_0.columns.map('_YEAR_'.join)

    #now we will merge those both agrregared daatframes bureau_balance_aggregate,
    agg_bureau_balance= bureau_balance_aggregate.merge(bureau_balance_aggregate_year_0,on='SK_ID_BUREAU',how='inner')
    #to save memory space
    del bureau_balance_aggregate
    del bureau_balance_aggregate_year_0
    #finally whahatever is left as NA we will fill it simply by 0

    #reseting index
    agg_bureau_balance=agg_bureau_balance.reset_index()
    agg_bureau_balance.fillna(0,inplace=True)
    
    #agg_bureau_balance.column=agg_bureau_balance.columns= ["bbl_"+col.upper() if col not in ["SK_ID_BUREAU"] else col for col in df.columns]


    return agg_bureau_balance 
    
```

```python
def preprocess_buerau(bureau):
    df=bureau.copy()
#     le= LabelEncoder()
#     le.fit(df['CREDIT_ACTIVE'])
#     df['CREDIT_ACTIVE']=le.transform(df['CREDIT_ACTIVE'])
    df['DAYS_CREDIT']=-1*df['DAYS_CREDIT']
    #Encoding the categorical columns in one-hot form
    currency_ohe = pd.get_dummies(df['CREDIT_CURRENCY'], prefix = 'CURRENCY')
    credit_active_ohe = pd.get_dummies(df['CREDIT_ACTIVE'], prefix = 'CREDIT_ACTIVE')
    credit_type_ohe = pd.get_dummies(df['CREDIT_TYPE'], prefix = 'CREDIT_TYPE')
    #CREDIT_DAY_OVERDUE
    df['CREDIT_DAY_OVERDUE']=-1*df['CREDIT_DAY_OVERDUE']
    #we are keeping the data which have DAYS_CREDIT_ENDDATE more than 50 years ie. 18250 as error as NAN
    df[df['DAYS_CREDIT_ENDDATE']> 18250]['DAYS_CREDIT_ENDDATE']=np.nan
    #Similiary we have seen the same for DAYS_ENDDATE_FACT,DAYS_CREDIT_UPDATE
    #we will fill na as same as DAYS_CREDIT_ENDDATE
    df[df['DAYS_ENDDATE_FACT']> 18250]['DAYS_ENDDATE_FACT']=np.nan
    df[df['DAYS_CREDIT_UPDATE']> 18250]['DAYS_CREDIT_UPDATE']=np.nan
    #now we will create some features based on loan data columns 
    df['Duration_of_CREDIT']= abs(abs(df['DAYS_CREDIT']) - abs((df['DAYS_CREDIT_ENDDATE'])))
    df["OVERDUE_ANNUITY_RATIO"] = df["AMT_CREDIT_SUM_OVERDUE"].fillna(0.01) / df["AMT_ANNUITY"].fillna(0.01)
    df["OVERDUE_CREDIT_RATIO"] = df["AMT_CREDIT_SUM_OVERDUE"].fillna(0.01) / df["AMT_CREDIT_SUM"].fillna(0.01)
    df["MAX_OVERDUE_ANNUITY_RATIO"] = df["AMT_CREDIT_MAX_OVERDUE"].fillna(0.01) / df["AMT_ANNUITY"].fillna(0.01)
    df["MAX_OVERDUE_CREDIT_SUM_RATIO"] = df["AMT_CREDIT_MAX_OVERDUE"].fillna(0.01) / df["AMT_CREDIT_SUM"].fillna(0.01)
    df["CREDIT_SUM_DEBT_LIMIT_RATIO"]    = df["AMT_CREDIT_SUM_DEBT"].fillna(0.01) / df["AMT_CREDIT_SUM_LIMIT"].fillna(0.01)
    df['SUM_OVERDUE_DAYS_CREDIT_RATIO']= df["AMT_CREDIT_SUM_OVERDUE"].fillna(0.01) / df["DAYS_CREDIT"].fillna(0.01)
    df["MAX_OVERDUE_DAYS_CREDIT_RATIO"] = df["AMT_CREDIT_MAX_OVERDUE"].fillna(0.01) / df["DAYS_CREDIT"].fillna(0.01)
    df["SUM_DEBT_DAYS_CREDIT_RATIO"]    = df["AMT_CREDIT_SUM_DEBT"].fillna(0.01) / df["DAYS_CREDIT"].fillna(0.01)
    
    df['DEBT_PER']=df['AMT_CREDIT_SUM_DEBT']/df['AMT_CREDIT_SUM']
    
    df['AMT_ANNUITY_AMT_CREDIT_SUM_PER']=df['AMT_ANNUITY']/df['AMT_CREDIT_SUM']
    
    df['DEBT_LIMIT_PER']=df['AMT_CREDIT_SUM_DEBT']/df['AMT_CREDIT_SUM_LIMIT']
    
    df['AMT_EXTRA_PAY'] = df['AMT_ANNUITY']-df['AMT_CREDIT_SUM']
    
    #Debt and credit ratio
    df["DEBT_CREDIT_RATIO"]    = df["AMT_CREDIT_SUM_DEBT"].fillna(0.01) / df["AMT_CREDIT_SUM"].fillna(0.01)
    #Secure and non secure loan type 
    # The loan which usually more secure and tend to be very less deafulter are
    #Car loan,Loan for the purchase of equipment,Mortgage,Real estate loan,Loan for purchase of shares (margin lending)
    # loans which usually less secure and tend to be more deafulter are
    #Another type of loan,Cash loan (non-earmarked), Consumer credit,Credit card, Interbank credit,Loan for business development,
    #Loan for working capital replenishment, Microloan, Mobile operator loan,Unknown type of loan
    # we will create a FLAG for that
    secure_loans=['Car loan','Loan for the purchase of equipment','Mortgage,Real estate loan','Loan for purchase of shares (margin lending)']
    unsecure_loan= ['Another type of loan','Cash loan (non-earmarked)', 'Consumer credit','Credit card', 
                    'Interbank credit','Loan for business development','Loan for working capital replenishment',
                    'Microloan, Mobile operator loan','Unknown type of loan']
    
    #lest create a Flag columns for that
    df['FLAG_SECURED_LOAN']= df['CREDIT_TYPE'].apply(lambda x : 1 if x in secure_loans else 0)
    df['FLAG_UNSECURED_LOAN']= df['CREDIT_TYPE'].apply(lambda x : 1 if x in unsecure_loan else 0)
    
    
    
    #now we will drop the credit _Type and Currency columns and add the one hot encoded feature for those categories
    df= pd.concat([df.drop(['CREDIT_CURRENCY','CREDIT_TYPE','CREDIT_ACTIVE'],axis=1),currency_ohe,credit_type_ohe,credit_active_ohe],axis=1)
    
    return df
```

```python
def bureau_merged_feature_engineering(bureau_merged):
    """
    Here after preprocessing and create features for bureau and bureau_balance_aggregated data,Now we have
    Megred data for bureau and bureau_balance_aggregate, we will group data by SK_ID_CURR and again create some interesting feature which might
    work well,will check it later
    """
    bureau_merged_agg={}
    
    for col in list(bureau_merged.columns):
        if (col!='SK_ID_CURR' or col!='SK_ID_BUREAU'):
            bureau_merged_agg[col]=['mean']
            if (col=='AMT_CREDIT_SUM_DEBT') | (col=='AMT_CREDIT_SUM_OVERDUE') | (col=='FLAG_SECURED_LOAN') |(col=='FLAG_UNSECURED_LOAN'):
                bureau_merged_agg[col]=['sum']
            if col=='DAYS_CREDIT':
                bureau_merged_agg[col]=['min','max','mean']
            if col=='DEBT_PER':
                bureau_merged_agg[col]=['mean']
                
    #Now this is time to agreegate the bureau+burealu balance_agg data to group by SK_ID_CURR and get the agrregate values
    bureau_merged_agg = bureau_merged.groupby('SK_ID_CURR').agg(bureau_merged_agg)
    bureau_merged_agg.columns= bureau_merged_agg.columns.map('_'.join)
     #reseting index
    bureau_merged_agg=bureau_merged_agg.reset_index()
    
    #getting the days credit year
    bureau_merged_agg['YEAR_CREDIT_min']=np.round((bureau_merged_agg['DAYS_CREDIT_min']/365),3)
    bureau_merged_agg['YEAR_CREDIT_max']=np.round((bureau_merged_agg['DAYS_CREDIT_max']/365),3)
    bureau_merged_agg['YEAR_CREDIT_mean']=np.round((bureau_merged_agg['DAYS_CREDIT_mean']/365),3)
    #Some more aggregate features for SK_ID_CURR
    bureau_merged_agg['AMT_CREDIT_MAX_OVERDUE_MAX'] = bureau_merged.groupby('SK_ID_CURR')['AMT_CREDIT_MAX_OVERDUE'].max()
    #-----------------------------------------------------------------------------------------------------------
    #we will try to agreegate some columns based on CREDIT_ACTIVE type of "Active" and "Closed" and with respect to 'SK_ID_CURR'

    
    
    bureau_merged_agg_active=bureau_merged[bureau_merged['CREDIT_ACTIVE_Active']==1].groupby('SK_ID_CURR').aggregate({'AMT_CREDIT_SUM':['min','max','mean'],
                                                                                                   'AMT_CREDIT_SUM_OVERDUE':['min','max','mean'],
                                                                                                   'AMT_CREDIT_SUM_DEBT':['min','max','mean'],
                                                                                                   'AMT_CREDIT_MAX_OVERDUE':['min','max','mean'],
                                                                                                   'AMT_CREDIT_SUM_LIMIT':['min','max','mean'],
                                                                                                    'SK_ID_BUREAU':['count']} )                      
    
    bureau_merged_agg_active.columns= bureau_merged_agg_active.columns.map('_Active_'.join)
    bureau_merged_agg_active=bureau_merged_agg_active.reset_index()
    #bureau_merged_agg_active['COUNT_ACTIVE']= bureau_merged_agg_active

    bureau_merged_agg_closed=bureau_merged[bureau_merged['CREDIT_ACTIVE_Closed']==1].groupby('SK_ID_CURR').aggregate({'AMT_CREDIT_SUM':['min','max','mean'],
                                                                                                   'AMT_CREDIT_SUM_OVERDUE':['min','max','mean'],
                                                                                                   'AMT_CREDIT_SUM_DEBT':['min','max','mean'],
                                                                                                   'AMT_CREDIT_MAX_OVERDUE':['min','max','mean'],
                                                                                                   'AMT_CREDIT_SUM_LIMIT':['min','max','mean'],
                                                                                                    'SK_ID_BUREAU':['count']})                      
    
    bureau_merged_agg_closed.columns= bureau_merged_agg_closed.columns.map('_Closed_'.join)
    bureau_merged_agg_closed= bureau_merged_agg_closed.reset_index()
    ##
    
    #Now we will merge this  Active and close data to the bureau_merged_agg
    bureau_merged_agg= bureau_merged_agg.merge(bureau_merged_agg_active,how='left',on=['SK_ID_CURR'])
    bureau_merged_agg= bureau_merged_agg.merge(bureau_merged_agg_closed,how='left',on=['SK_ID_CURR'])
    
    bureau_merged_agg=bureau_merged_agg.drop(['SK_ID_CURR_mean','SK_ID_BUREAU_mean'],axis=1)
    bureau_merged_agg.columns= [col.upper() for col in bureau_merged_agg.columns]
    del bureau_merged_agg_closed
    del bureau_merged_agg_active
    #
    bureau_merged_agg.columns=bureau_merged_agg.columns= ["BBL_"+col.upper() if col not in ["SK_ID_CURR"] else col for col in bureau_merged_agg.columns]

    return bureau_merged_agg
```

```python

#import scipy
def preprocess_agg_fearure_eng_prev_app_data(df):
    
    #usually we should try to encode the Categorical columns first but what i have found is if we do it in start only our
    #grouping columns wont mean anything,Instead if we will try to do one hot encoding,It may increase the dimension but we will slice the
    #important features after fearure selection
    #So we will start with imputing outliers for some columns,we have seen in EDA
    continous_cols= prev_app_data.select_dtypes(include=['int64','float64']).columns.to_list()
    #in the EDA we found some erroneous values in DAYS columns, so we will replace them with NaN values
    for col in df.columns:
        if col.startswith('DAYS'):
            df[col].replace(365243, np.nan, inplace= True)
            
    
    #SELLERPLACE_AREA have a abrupt values as 4000000,we will fillNA for it
    df['SELLERPLACE_AREA'][df['SELLERPLACE_AREA'] == 4000000] = np.nan
    
    
    #############Some domain based features,WithOUT Grouping the data#######################################################
    #https://www.kaggle.com/c/home-credit-default-risk/discussion/64598# 10 place guy's best features
    df= df.drop("WEEKDAY_APPR_PROCESS_START",axis=1)
    df['AMT_INTEREST'] = df['CNT_PAYMENT']*df['AMT_ANNUITY']-df['AMT_CREDIT']
    df['AMT_ACTUAL_LOAN_AMT'] = df['CNT_PAYMENT']*df['AMT_ANNUITY']-df['AMT_DOWN_PAYMENT']
    #This could be a good feature that is interest rate,I havent seen interest rate in this data
    #we we are trying to create one.
    df['INTEREST_RATE'] = 2*12*df['AMT_INTEREST']/(df['AMT_CREDIT']*(df['CNT_PAYMENT']+1))
    df['INTEREST_SHARE'] = df['AMT_INTEREST']/df['AMT_CREDIT']
    #PRT*100,Usually it should be Compound intereset
    df['SIMPLE_INTEREST_PERYEAR']= (df['AMT_CREDIT']*df['INTEREST_RATE']*df['CNT_PAYMENT'])/100
    #Compound interest per year,CI = P( 1 + r/100)n - P
    #df['COMPOUND_INTEREST_PERYEAR']= (df['AMT_CREDIT']*((1+(df['RATE_OF_INTREST']/100))**(df['RATE_OF_INTREST']*df['CNT_PAYMENT'])))-df['AMT_CREDIT']
    df["AMT_CREDIT_APPLIED_RATIO"]  = df["AMT_CREDIT"] / df["AMT_APPLICATION"]
    df["AMT_GOODSPRICE_APPLIED_RATIO"]  = df["AMT_GOODS_PRICE"] / df["AMT_APPLICATION"]
    df["DOWN_PAYMENT_APPLIED_RATIO"] = df["AMT_DOWN_PAYMENT"] / df["AMT_APPLICATION"]
    #day differences
    df["DAYS_DUE_DIFF1"] = df["DAYS_LAST_DUE_1ST_VERSION"] - df["DAYS_FIRST_DUE"]
    df["DAYS_LAST_FIRST_DIFF2"] = df["DAYS_LAST_DUE"] - df["DAYS_FIRST_DUE"]
    df["DAYS_TERMINATION_FIRST_DRAWING_DIFF"] = df["DAYS_TERMINATION"] - df["DAYS_FIRST_DRAWING"]
    df["DAYS_TERMINATION_FIRST_DUE_DIFF"] = df["DAYS_TERMINATION"] - df["DAYS_FIRST_DUE"]
    df["DAYS_TERMINATION_LAST_DUE_DIFF"] = df["DAYS_TERMINATION"] - df["DAYS_LAST_DUE"]
    #Ratio for down pyament done and credited Amount
    df['CREDIT_DOWNPAYMENT_RATIO'] = df['AMT_DOWN_PAYMENT'] / (df['AMT_CREDIT'] + 0.001)
    df['GOOD_DOWNPAYMET_RATIO'] = df['AMT_DOWN_PAYMENT'] / (df['AMT_GOODS_PRICE'] + 0.001)
    df['RATE_DOWNPAYMENT'] = df['RATE_DOWN_PAYMENT'] * df['AMT_DOWN_PAYMENT']
    df['INTEREST_CREDIT'] = df['AMT_CREDIT'] * df['RATE_INTEREST_PRIMARY']
    df['INTEREST_CREDIT_PRIVILEGED'] = df['AMT_CREDIT'] * df['RATE_INTEREST_PRIVILEGED']
    df['APPLICATION_AMT_TO_DECISION_RATIO'] = df['AMT_APPLICATION'] / (df['DAYS_DECISION'] + 0.001) * -1
    df['AMT_APPLICATION_TO_SELLERPLACE_AREA'] = df['AMT_APPLICATION'] / (df['SELLERPLACE_AREA'] + 0.001)
    df['ANNUITY'] = df['AMT_CREDIT'] / (df['CNT_PAYMENT'] + 0.001)
    df['MISSING_VALUES_TOTAL_PREV'] = df.isna().sum(axis = 1)
    df['AMT_SURPLUS_DECLINED'] = df['AMT_APPLICATION'] - df['AMT_CREDIT']
    df['AMT_CREDIT_GOODS_RATIO'] = df['AMT_CREDIT'] / (df['AMT_GOODS_PRICE'] + 0.001)
    df['AMT_CREDIT_GOODS_DIFF'] = df['AMT_CREDIT'] - df['AMT_GOODS_PRICE']
    df['AMT_CREDIT_APPLICATION_RATIO'] = df['AMT_APPLICATION'] / (df['AMT_CREDIT']+0.001)
    df['ANNUITY_GOODS'] = df['AMT_GOODS_PRICE'] / (df['CNT_PAYMENT'] + 0.001)
    df['DAYS_FIRST_LAST_DUE_DIFF' ] = df['DAYS_LAST_DUE'] - df['DAYS_FIRST_DUE']
    df['AMT_CREDIT_NFLAG_LAST_APPL_DAY'] = df['AMT_CREDIT'] * df['NFLAG_LAST_APPL_IN_DAY']
    df['AMT_DOWN_PAYMENT_30']=(df['AMT_DOWN_PAYMENT']<=(0.30*df['AMT_CREDIT'])).astype(int)
    df['cnt_missing_vals']=df.isna().sum(axis = 1)
    ############## Prev approved/not approved count of the loan appliacation#############################
    cnt_prev_apps= df[["SK_ID_CURR", "SK_ID_PREV"]].groupby(["SK_ID_CURR"], as_index = False).count()
    cnt_prev_apps.columns = ["SK_ID_CURR", "CNT_PREV_APPLICATIONS"]
    df = df.merge(cnt_prev_apps, how = "left", on = "SK_ID_CURR")
    del cnt_prev_apps
    ##############################FLAG_LAST_APPL_PER_CONTRACT##########################################
    flag_last_app_per_contract_Y= df[df['FLAG_LAST_APPL_PER_CONTRACT']=='Y']
    cnt_flag_last_app_per_contract= flag_last_app_per_contract_Y[["SK_ID_CURR", "FLAG_LAST_APPL_PER_CONTRACT"]].groupby(["SK_ID_CURR"], as_index = False).count()
    cnt_flag_last_app_per_contract.columns=["SK_ID_CURR", "CNT_PREV_CONTRACTS"]
    df= df.merge(cnt_flag_last_app_per_contract,how='left',on = "SK_ID_CURR")
    del cnt_flag_last_app_per_contract
    ##Now we will get ratio of both
    df['RATIO_CNT_PREV_APP_CONTRACT']=  df["CNT_PREV_APPLICATIONS"] / df["CNT_PREV_CONTRACTS"]
    ##########NAME_CONTRACT_STATUS ###################################################
    #for "Approved"
    approved= df[df['NAME_CONTRACT_STATUS']=='Approved']
    appoved_group= df[["SK_ID_CURR", "NAME_CONTRACT_STATUS"]].groupby(["SK_ID_CURR"], as_index = False).count()
    appoved_group.columns = ["SK_ID_CURR", "CNT_CONTRACT_STATUS_APPROVED"]
    df = df.merge(appoved_group, how = "left", on = "SK_ID_CURR")
    del approved
    #For Refused
    not_approved= df[(df['NAME_CONTRACT_STATUS']=='Refused')|df['NAME_CONTRACT_STATUS']=='Canceled']
    not_approved_group= df[["SK_ID_CURR", "NAME_CONTRACT_STATUS"]].groupby(["SK_ID_CURR"], as_index = False).count()
    not_approved_group.columns = ["SK_ID_CURR", "CNT_CONTRACT_STATUS_REFUSED"]
    df = df.merge(not_approved_group, how = "left", on = "SK_ID_CURR")
    ##########################LOG Features##########################################################
    #we will take log of most amounts columns 
    amount_cols=  ["AMT_CREDIT", "AMT_ANNUITY", "AMT_APPLICATION", "AMT_DOWN_PAYMENT", "AMT_GOODS_PRICE"]
    for i in amount_cols:
        df["LOG_"+str(i)]=np.log(abs(df[i])+1)
    ######creating lag features,read from KAggle solutions:https://www.kaggle.com/c/home-credit-default-risk/discussion/64485
    ###So the idea is we will take  for each SK_ID_CURR,sort it by DAYS_DECISION we will take lag of NAME_CONTRACT_STATUS by 1st 3rd and 5th and 
    #get the ratio for mean of approved /rejected applications,
    # preparations for this process
    lags = [1, 3, 5]
    decision_prev = df[["SK_ID_CURR", "SK_ID_PREV", "DAYS_DECISION", "NAME_CONTRACT_STATUS"]]
    decision_prev["DAYS_DECISION"] = -decision_prev["DAYS_DECISION"]# just for sorting puporse
    decision_prev = decision_prev.sort_values(by = ["SK_ID_CURR","SK_ID_PREV", "DAYS_DECISION"])
    decision_prev = pd.get_dummies(decision_prev)

    # compuatation
    for l in lags:
        #acceptance ratios
        tmp = decision_prev[["SK_ID_CURR", "NAME_CONTRACT_STATUS_Approved"]].groupby(["SK_ID_CURR"]).head(l)
        tmp = tmp.groupby(["SK_ID_CURR"], as_index = False).mean()
        tmp.columns = ["SK_ID_CURR", "CONTRACT_STATUS_APPROVE_MEAN_" + str(l)]
        df = df.merge(tmp, how = "left", on = "SK_ID_CURR")

    # rejection ratios
    tmp = decision_prev[["SK_ID_CURR", "NAME_CONTRACT_STATUS_Refused"]].groupby(["SK_ID_CURR"]).head(l)
    tmp = tmp.groupby(["SK_ID_CURR"], as_index = False).mean()
    tmp.columns = ["SK_ID_CURR", "CONTRACT_STATUS_REJECT_MEAN_" + str(l)]
    df = df.merge(tmp, how = "left", on = "SK_ID_CURR")
    
    del tmp,decision_prev
    #
    #### one hot encoding for the Categorical columns############################################
    df = pd.get_dummies(df)
    #return df
    #####################Agreegation Features####################################################
    prev_data_agg={}
    for col in df.columns:
        if col!='SK_ID_CURR' and col !='SK_ID_PREV':
            prev_data_agg[col]=['mean']
        if (col in ['DAYS_TERMINATION','DAYS_FIRST_DUE''DAYS_LAST_DUE','AMT_CREDIT',
                    'AMT_ANNUITY','AMT_DOWN_PAYMENT','DAYS_LAST_DUE_1ST_VERSION']):
            prev_data_agg[col]=['min','max','mean','sum']
            
    df_agg = df.groupby('SK_ID_CURR').agg(prev_data_agg)
    df_agg.columns= df_agg.columns.map('_'.join)
     #reseting index
    df_agg=df_agg.reset_index()
    ############################################################################################
    #Now as we have done in bureau data,we will slice the data into NAME_CONTRACT_STATUS
    #into "Approved" and "Cancelled/Refused " and create some agrregate features
    features_agg_col={'AMT_CREDIT':['min','max','mean','sum'],
                      'AMT_ANNUITY':['min','max','mean','sum'],
                      #'AMT_CREDIT_SUM_DEBT':['min','max','mean','sum'],
                      'AMT_DOWN_PAYMENT':['min','max','mean','sum'],
                      'AMT_ACTUAL_LOAN_AMT':['min','max','mean'],
                      'INTEREST_RATE':['min','max','mean','sum'],
                      'SK_ID_PREV':['count']}
    #1.Approved Data
    approved_agg = df[(df['NAME_CONTRACT_STATUS_Approved'] == 1)].groupby('SK_ID_CURR').aggregate(features_agg_col)
    approved_agg.columns= approved_agg.columns.map('_prev_Approved_'.join)
    approved_agg=approved_agg.reset_index()
    #1.Not approved Data
    not_approved = df[(df['NAME_CONTRACT_STATUS_Refused'] == 1) | (df['NAME_CONTRACT_STATUS_Canceled'] == 1)]
    
    not_approved_agg=not_approved.groupby('SK_ID_CURR').aggregate(features_agg_col)                      
    
    not_approved_agg.columns= not_approved_agg.columns.map('_prev_NOT_Approved_'.join)
    not_approved_agg=not_approved_agg.reset_index()
    #Now we will merge this  Approved and not approved data to the bureau_merged_agg
    df_agg= df_agg.merge(approved_agg,how='left',on=['SK_ID_CURR'])
    df_agg= df_agg.merge(not_approved_agg,how='left',on=['SK_ID_CURR'])
    
    #bureau_merged_agg=bureau_merged_agg.drop(['SK_ID_CURR_mean','SK_ID_BUREAU_mean'],axis=1)
    #df_agg.columns= [col.upper() for col in df_agg.columns]
    df_agg.columns=df_agg.columns= ["PREVAPP_"+col.upper() if col not in ["SK_ID_CURR"] else col for col in df_agg.columns]

    
    del not_approved_agg,not_approved,approved_agg
    return df_agg
```

# POS_CASH_balance.csv

- This table contains Monthly balance of client's previous credit loans in Home Credit along with behavioral data
- Monthly balance snapshots of previous POS (point of sales) and cash loans that the applicant had with Home Credit.
- This table has one row for each month of history of every previous credit in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample # of relative previous credits # of months in which we have some history observable for the previous credits) rows.

```python
def preprocess_agg_feature_engineering_pos_balance(df):
    
    df['MONTHS_BALANCE']= -1*df['MONTHS_BALANCE']
    df["CNT_INSTALLMENT_PAID"]= df['CNT_INSTALMENT']- df['CNT_INSTALMENT_FUTURE']
    #leave data where NAME_CONTRACT_STATUS is XNA
    df=df[df['NAME_CONTRACT_STATUS']!='XNA']
    df['SK_DPD_RATIO'] = df['SK_DPD'] / (df['SK_DPD_DEF'] + 0.0001)
    #df['TOTAL_INSTALLMENTS'] = df['CNT_INSTALMENT'] + df['CNT_INSTALMENT_FUTURE']
       
    #one hot encoding NAME_CONTRACT_STATUS
    df =pd.get_dummies(df)
    #agreegating data
    pos_agg_dict={}
    for col in df.columns:
        if col!='SK_ID_CURR' and col !='SK_ID_PREV':
            pos_agg_dict[col]=['mean','min','max']
        if col =='MONTHS_BALANCE':
            pos_agg_dict[col]=['max','mean','size','sum']
    #Apply Aggregation
    df_agg = df.groupby('SK_ID_CURR').agg(pos_agg_dict)
    df_agg.columns= df_agg.columns.map('_'.join)
    df_agg['pos_count']= df.groupby('SK_ID_CURR')['SK_ID_PREV'].size()
    
     #reseting index
    df_agg=df_agg.reset_index()
    df_agg.columns= ['SK_ID_CURR']+[str(col).upper() for col in df_agg.columns if col !='SK_ID_CURR']
    #we will take diff between NAME_CONTRACT_STATUS=="Active" and NAME_CONTRACT_STATUS=="Completd"
    df_agg['ACTIVE_COMPLETED_MEAN_DIFF']=df_agg['NAME_CONTRACT_STATUS_COMPLETED_MEAN']-df_agg['NAME_CONTRACT_STATUS_ACTIVE_MEAN']
    df_agg['ACTIVE_COMPLETED_MIN_DIFF']=df_agg['NAME_CONTRACT_STATUS_COMPLETED_MIN']-df_agg['NAME_CONTRACT_STATUS_ACTIVE_MIN']
    df_agg['ACTIVE_COMPLETED_MAX_DIFF']=df_agg['NAME_CONTRACT_STATUS_COMPLETED_MAX']-df_agg['NAME_CONTRACT_STATUS_ACTIVE_MAX']
    
    #we will take the NAME_CONTRACT_STATUS
    #Now as we have done in bureau data,we will slice the data into NAME_CONTRACT_STATUS
    #into "Approved" and "Cancelled/Refused " and create some agrregate features
    features_agg_col={'CNT_INSTALMENT':['min','max','mean','sum'],
                      'CNT_INSTALMENT_FUTURE':['min','max','mean'],
                      #'AMT_CREDIT_SUM_DEBT':['min','max','mean','sum'],
                      'SK_DPD':['min','max','mean','sum'],
                      'SK_DPD_DEF':['min','max','mean'],
                      #'INTEREST_RATE':['min','max','mean','sum'],
                      'SK_ID_PREV':['size']}
    #1.Active Data
    Active_agg = df[(df['NAME_CONTRACT_STATUS_Active'] == 1)].groupby('SK_ID_CURR').aggregate(features_agg_col)
    Active_agg.columns= Active_agg.columns.map('_CONTRACT_ST_ACTIVE_'.join)
    Active_agg=Active_agg.reset_index()
    #1.CLosed Data
    Closed_agg = df[(df['NAME_CONTRACT_STATUS_Completed'] == 1)].groupby('SK_ID_CURR').aggregate(features_agg_col)                      
    
    Closed_agg.columns= Closed_agg.columns.map('_CONTRACT_ST_CLOSED_'.join)
    Closed_agg=Closed_agg.reset_index()
    #Now we will merge this  Approved and not approved data to the bureau_merged_agg
    df_agg= df_agg.merge(Active_agg,how='left',on=['SK_ID_CURR'])
    df_agg= df_agg.merge(Closed_agg,how='left',on=['SK_ID_CURR'])
    df_agg.columns= ['SK_ID_CURR']+['POS_'+ str(col).upper() for col in df_agg.columns if col !='SK_ID_CURR']

    #https://www.kaggle.com/c/home-credit-default-risk/discussion/64598
    #from the kaggle 10th place solution
    
    
    
    return df_agg
    
```

```python
def preprocess_agg_feature_engineering_cc_balance(df):
    
    df['MONTHS_BALANCE']= abs(df['MONTHS_BALANCE'])
    #We have seen the data is for 8 year and monthwise -1 was the freshest months and as it
    #decrease(more negative) it getting older
    #removing outliers in AMT_PAYMENT_CURRENT
    df[df['AMT_PAYMENT_CURRENT'] > 4000000]['AMT_PAYMENT_CURRENT'] = np.nan
    #we see so many columns as "DRAWING" this is for windrawl,we can sum it
    df['AMT_DRAWING_SUM'] = df['AMT_DRAWINGS_ATM_CURRENT'].fillna(0) + df['AMT_DRAWINGS_CURRENT'].fillna(0)+df['AMT_DRAWINGS_OTHER_CURRENT'].fillna(0) + df['AMT_DRAWINGS_POS_CURRENT'].fillna(0)
    df['CNT_DRAWING_SUM'] = df['CNT_DRAWINGS_ATM_CURRENT'].fillna(0) + df['CNT_DRAWINGS_CURRENT'].fillna(0)+df['CNT_DRAWINGS_OTHER_CURRENT'].fillna(0) + df['CNT_DRAWINGS_POS_CURRENT'].fillna(0)
    #similarly we can see AMT_RECIEVALE
    df['AMT_RECEIVABLE_SUM'] = df['AMT_RECEIVABLE_PRINCIPAL'].fillna(0) + df['AMT_RECIVABLE'].fillna(0)+df['AMT_TOTAL_RECEIVABLE'].fillna(0)
    #how much of percentage the total card limit is used
    df['LIMIT_USED_PER']= df['AMT_BALANCE']/(df['AMT_CREDIT_LIMIT_ACTUAL']+1e-4)
    #How much LIMIT left??
    df['LIMIT_USED_LEFT']= df['AMT_CREDIT_LIMIT_ACTUAL'].fillna(0)+df['AMT_BALANCE'].fillna(0)
    #Irregularity in payment
    df['IRRULARITY_IN_PAYMENT']= df['AMT_PAYMENT_CURRENT']/(df['AMT_INST_MIN_REGULARITY']+1e-4)
    #df['AMT_LEFT_TO_PAY']= df['AMT_PAYMENT_CURRENT'].fillna(0)-(df['AMT_INST_MIN_REGULARITY'].fillna(0))
    df["DPD_RATIO"]= df['SK_DPD']/(df['SK_DPD_DEF']+1e-4)
    #Interest is total amount to pay - Principal
    df['INTERST_RECIEVABLE']= df['AMT_TOTAL_RECEIVABLE'] - df['AMT_RECEIVABLE_PRINCIPAL']
    #let FLAG the data where the 25% ,50%,75% and 95% credit limit is used
    df['FLAG_LIMIT_USED_25_PER']= (df['AMT_DRAWING_SUM']<(0.25*(df['AMT_CREDIT_LIMIT_ACTUAL']))).astype(int)
    df['FLAG_LIMIT_USED_50_PER']= (df['AMT_DRAWING_SUM']<(0.50*(df['AMT_CREDIT_LIMIT_ACTUAL']))).astype(int)
    df['FLAG_LIMIT_USED_75_PER']= (df['AMT_DRAWING_SUM']<(0.75*(df['AMT_CREDIT_LIMIT_ACTUAL']))).astype(int)
    df['FLAG_LIMIT_USED_95_PER']= (df['AMT_DRAWING_SUM']<(0.95*(df['AMT_CREDIT_LIMIT_ACTUAL']))).astype(int)
    df['FLAG_OVERLIMIT']=(df['AMT_DRAWING_SUM']>df['AMT_CREDIT_LIMIT_ACTUAL']).astype(int) 
    #One hot encoding for NAME_CONTRACT_STATUS##############
    df=pd.get_dummies(df)
#     log_vars = ["AMT_BALANCE", "AMT_CREDIT_LIMIT_ACTUAL", "AMT_DRAWINGS_ATM_CURRENT", "AMT_DRAWINGS_CURRENT",
#             "AMT_DRAWINGS_OTHER_CURRENT", "AMT_DRAWINGS_POS_CURRENT", "AMT_INST_MIN_REGULARITY",
#             "AMT_PAYMENT_CURRENT", "AMT_PAYMENT_TOTAL_CURRENT", "AMT_RECEIVABLE_PRINCIPAL",
#             "AMT_RECIVABLE", "AMT_TOTAL_RECEIVABLE"]
# #     for col in log_vars:
#         df[col]= np.log(abs(df[col])+1)
# Log variable didnt come with better AUC,So we didnt use it

    #There is one interseting thing in this table we found in EDA
    #we have total 104307 SK_ID_PREV and 103558 SK_ID_CURR
    #This mean mostl SK_ID_PREV are same and value is for each month for credit card,
    #from All customer applied for loan,not everyone wont have credit cards 
    #well its very correct ,SO we need to get the aggregate data first "SK_ID_PREV" and then "SK_ID_CURR"
    
    #Agreegate on SK_ID_PREV,we will just take mean and sum for FLAG_cols
#     cc_bal_agg={}
#     #cc_bal_agg['SK_ID_CURR']=['first']

#     for col in df.columns:
#         if col!="SK_ID_PREV" and col!='SK_ID_CURR':
#             cc_bal_agg[col]=['mean']
#             if str(col).startswith("FLAG") or str(col).startswith("NAME_CONTRACT_STATUS"):
#                 cc_bal_agg[col]=['sum','mean']
#             if col=='MONTH_BALANCE':
#                 cc_bal_agg[col]=['min','mean']
                
    
#     cc_bal_agg_PREV= df.groupby(['SK_ID_PREV','SK_ID_CURR']).agg(cc_bal_agg)
#     cc_bal_agg_PREV.columns= cc_bal_agg_PREV.columns.map('_'.join)
#     cc_bal_agg_PREV=cc_bal_agg_PREV.reset_index()
######**Actually i tried this ,but things got complicated only,SO i will directly group by SK_ID_CURR####
#   Now we will group on SK_ID_CURR,this wont have so much differnece from the prev table
    cc_bal_agg={}
    #cc_bal_agg['SK_ID_CURR']=['first']

    for col in df.columns:
        if col!="SK_ID_PREV" and col!='SK_ID_CURR':
            cc_bal_agg[col]=['mean']
            if str(col).startswith("FLAG") or str(col).startswith("NAME_CONTRACT_STATUS"):
                cc_bal_agg[col]=['sum','mean']
            if col=='MONTHS_BALANCE':
                cc_bal_agg[col]=['min','mean','max']
                
    #cc_bal_agg['SK_ID_PREV']=['size']
    
    
    cc_bal_agg_CURR= df.groupby(['SK_ID_CURR']).agg(cc_bal_agg)
    cc_bal_agg_CURR.columns= cc_bal_agg_CURR.columns.map('_'.join)
    cc_bal_agg_CURR=cc_bal_agg_CURR.reset_index()
    cc_bal_agg_CURR.columns=cc_bal_agg_CURR.columns= ["CCBAL_"+col.upper() if col not in ["SK_ID_CURR"] else col for col in cc_bal_agg_CURR.columns]

    del df
    gc.collect()
    
    return cc_bal_agg_CURR
```

# installments_payments.csv

This table consist past payment data for each installment of previous credits in Home credit realted to loans in our sample

```python
def preprocess_agg_feature_engineering_install_payments(df):
    
    #domain based features before aggregation of data
    df['PAYMENT_INSTALLEMENT_NUM_DIFF']=df['NUM_INSTALMENT_VERSION']-df['NUM_INSTALMENT_NUMBER']
    df['DAYS_LATE_PAYMENT']= abs(df['DAYS_ENTRY_PAYMENT'].fillna(0))- abs(df['DAYS_INSTALMENT'])
    df['INSTALLMENT_PAYMENT_DIFF']= df['AMT_INSTALMENT']-df['AMT_PAYMENT'].fillna(0)
    #we have seen the DAYS_ENTRY_PAYMENT and AMT_PAYMENT are NA for the same row
    #Its obvious actually if amount is not paid so there will no DAYS_ENTRY_PAYMENT
    #we can create a flag for it
    df['FLAG_NO_PAYMENT']=(df['AMT_PAYMENT'].isna() & df['DAYS_ENTRY_PAYMENT'].isna())
    #we can create a flag for late payment
    df['FLAG_LATE_PAYMENT']= ((abs(df['DAYS_ENTRY_PAYMENT'].fillna(0))- abs(df['DAYS_INSTALMENT']))>1).astype(int)
    #we can also create a flag if there is a different in amount paid
    df['FLAG_LESS_PAYMENT']= ((abs(df['AMT_INSTALMENT'].fillna(0))- abs(df['AMT_PAYMENT']))>1).astype(int)   
    #For NUM_INSTALMENT_VERSION we have seen most values are just 1,2,0 (descending order)
    # we have tried try to group it by "SK_ID_PREV" but didnt helped,So we will try to group it by
    #'SK_ID_CURR'
    install_pay_agg={}
    for col in df.columns:
        if col!='SK_ID_CURR' and col !='SK_ID_PREV':
            install_pay_agg[col]=['mean','max','sum']
        if str(col).startswith('FLAG'):
            install_pay_agg[col]=['sum']
        if (col=='PAYMENT_INSTALLEMENT_NUM_DIFF')|(col=='DAYS_LATE_PAYMENT')|(col=='INSTALLMENT_PAYMENT_DIFF'):
            install_pay_agg[col]=['mean','max','sum']
        if col=='SK_ID_PREV':
            install_pay_agg[col]=['size']
            
    df_agg = df.groupby('SK_ID_CURR').agg(install_pay_agg)
    df_agg.columns= df_agg.columns.map('_'.join)
    df_agg=df_agg.reset_index()
    df_agg.columns=df_agg.columns= ["INSTLPAY_"+col.upper() if col not in ["SK_ID_CURR"] else col for col in df_agg.columns]

    del df
    gc.collect()
    return df_agg
            
```
