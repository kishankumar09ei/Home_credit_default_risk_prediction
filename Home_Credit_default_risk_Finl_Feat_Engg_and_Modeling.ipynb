<h2 style = "color : Black"><center>Business Problem </center> </h2>

The Lending Organizations are faced with the tough task of analyzing the risk associated with each client. Therefore, it is important to identify the risky behaviors of clients and make educated decisions. The two most critical questions in the banking industry are 
- (i) How risky is the borrower? 

- (ii) Given the borrower’s risk, should we lend him/her?

Many people struggle to get loans due to insufficient or non-existent credit histories. And, unfortunately, this population is often taken advantage of by untrustworthy lenders. Here we face issue of CREDIT RISK.

Default Credit risk is the chance that companies or individuals will be unable to make the required payments on their debt obligations. In other words, credit default risk is the probability that if you lend money, there is a chance that they won’t be able to give the money back on time. Lenders and investors are exposed to default risk in virtually all forms of credit extensions. To mitigate the impact of default risk, lenders often impose charges that correspond to the debtor’s level of default risk. A higher level of risk leads to a higher required return. 

Many people struggle to get loans due to insufficient or non-existent credit histories. And, unfortunately, this population is often taken advantage of by untrustworthy lenders. Doing so will ensure that clients capable of repayment are not rejected and that loans are given with a principal, maturity, and repayment calendar that will empower their clients to be successful.
So, Lending Organizations are approached usually by TWO types of customers:

- 1.	People who do not particularly have a prior credit history, for example students’ loans, small businessmen, etc. who need credits, be it for studies, or for setting up some sort of businesses. Without adequate credit history, the lending organizations find it difficult to lend credits to such people, as these loans could be associated with high risks. In these kinds of situations, some lending organizations even tend to exploit the borrowers by asking for too high of an interest rate.
- 2.	People, who do have prior credit history, which could be with the same organization or some other organizations. However, going through that historical data could be very time consuming and redundant. This would scale up even further as the number of applicants increases. For such cases, if there could be a way through which the lending organization could predict or estimate the borrower’s repayment capability, the process could be streamlined and be made effective for both the lender and the borrower. It could save resources both in terms of humans and time.

**application_{train|test}.csv:**

 - This is the main table, broken into two files for Train (with TARGET) and Test (without TARGET).
 -  Static data for all applications. One row represents one loan in our data sample.

**bureau.csv :**

All client's previous credits provided by other financial institutions that were reported to Credit Bureau (for clients who have a loan in our sample).For every loan in our sample, there are as many rows as number of credits the client had in Credit Bureau before the application date.

**bureau_balance.csv:**

- Monthly balances of previous credits in Credit Bureau.
- This table has one row for each month of history of every previous credit reported to Credit Bureau – i.e the table has (#loans in sample * # of relative previous credits * # of months where we have some history observable for the previous credits) rows.

**POS_CASH_balance.csv:**

- Monthly balance snapshots of previous POS (point of sales) and cash loans that the applicant had with Home Credit.
- This table has one row for each month of history of every previous credit in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample * # of relative previous credits * # of months in which we have some history observable for the previous credits) rows.

**credit_card_balance.csv:**

- Monthly balance snapshots of previous credit cards that the applicant has with Home Credit.
- This table has one row for each month of history of every previous credit in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample * # of relative previous credit cards * # of months where we have some history observable for the previous credit card) rows.

**previous_application.csv:**

- All previous applications for Home Credit loans of clients who have loans in our sample.
- There is one row for each previous application related to loans in our data sample.

**installments_payments.csv:**

- Repayment history for the previously disbursed credits in Home Credit related to the loans in our sample.
- There is a) one row for every payment that was made plus b) one row each for missed payment.
- One row is equivalent to one payment of one instalment OR one instalment corresponding to one payment of one previous Home Credit related to loans in our sample.

**HomeCredit_columns_description.csv:**

- This file contains descriptions for the columns in the various data files.
- Our interest of data is application_{train|test}.csv
- We can get column description of each column in application_train.csv from HomeCredit_columns_description.csv.

```python
# WE WILL TRY TO CREATE OUR OWN RECURSIVE FEATURE SELECTION WAY AND IMPLEMENT IT
from datetime import datetime
def recursive_feature_selection_with_CV(X_train,y_train,cv):
    

    start= datetime.now()
    
    imp_cols=set()
    score=1
    i=1
    
    while score>0.78:
        
        print(f"{i} th iteration running")
        
        current_data= X_train.drop(list(imp_cols),axis=1)
        
        #define the CV 
        
        fold= StratifiedKFold(n_splits=cv,shuffle=True,random_state=33)
        
        ##again initializing score
        score=0
        feature_importance= np.zeros_like(current_data.columns)
        
        #doing k-fold Cross validation
        
        for fold_k,(train_index,val_index) in enumerate(fold.split(current_data,y_train),1):#because we want to index to start with 1
            
            #define the train and validation data
            print("\t \t Starting K fold validation of fold number : ",fold_k)
            X_train_cv=current_data.iloc[train_index]
            y_train_cv=y_train.iloc[train_index]
            X_val_cv=current_data.iloc[val_index]
            y_val_cv=y_train.iloc[val_index]
            
            #####Intializing LIGHTGBM classifier and fit train data
            lgbm_clf = LGBMClassifier(n_jobs=-1,random_state=33)
            lgbm_clf.fit(X_train_cv,y_train_cv)
            
            #after training we get feature importance of each feature ; we take its average over different folds, say K-Folds
            
            feature_importance += lgbm_clf.feature_importances_/cv
            
            #average k-fold ROC-AUC Score
            score += roc_auc_score(y_val_cv,lgbm_clf.predict_proba(X_val_cv)[:,1])/ cv
            
        #getting important columns
        #For safer side we will keep the features which have feature_importance >0
        imp_col_indexes= np.where(abs(feature_importance)>0)
        
        #Getting important columns
        
        imp_cols_fold_k= X_train.columns[imp_col_indexes]
        
        ## so if score is more than 0.7 than we can add/ update this to real imp_cols set so we can keep or remove features
        ##for next iteration,till our ovreall_score is good
        
        
        if score >0.75:
            imp_cols.update(imp_cols_fold_k)
            print(f"\t for iteration {i} total important columns are {len(imp_cols)}\ n")
        #print the score 
        print(f"\t AVG CV score with {len(imp_cols)} is {score} ")
        i +=1
    
    ### after all epoch run with good score we will return important_columns only
    print(f"After Recursive feature Selection ,Out of Total: {len(X_train.columns)} features; we have removed{len(X_train.columns)-len(imp_cols)} features")
    print("Time Taken",(datetime.now()-start))
    return list(imp_cols)
```

```python
# This function plots the confusion matrices given y_i, y_i_hat.
# This is not like a simple Confusion matrix but this will have all 3:  confusion matrix ,Precision matrix , Recall matrix

def plot_confusion_matrices(test_y,y_pred):
    C= confusion_matrix(test_y,y_pred)
    #The above one is the normal confusion matrix, a 2*2 matrix as[[TP,FP],
                                                                  #[[FN,TN]] 
    #Now we will get a Recall matrix, Recall is  TP/(Total positive pts = TP+FN),SO recall is how good we have predited
    #the Postive points correctly ,So we can divide the each element by the denominator ie. 
    # 1.(C.T / C.sum(axis=1)).T
    # C = [[1, 2],
    #     [3, 4]]
    # C.T = [[1, 3],
    #        [2, 4]]
    # C.sum(axis = 1)  axis=0 corresonds to columns and axis=1 corresponds to rows in two diamensional array
    # C.sum(axix =1) = [[3, 7]]
    # ((C.T)/(C.sum(axis=1))) = [[1/3, 3/7]
    #                           [2/3, 4/7]]

    # ((C.T)/(C.sum(axis=1))).T = [[1/3, 2/3]
    #                           [3/7, 4/7]]
    # sum of row elements = 1
    
    
    A = (((C.T)/(C.sum(axis=1))).T)
    #A=  np.array(A,dtype= np.int32)
    # recall matrix
    #Now we can get the Precision matrix,
    #Precision is How well we have identified point correctly in true and predicted of that class
    # Precision = TP/(Tp+FP)
    B= (C / C.sum(axis=0))
    
    plt.figure(figsize=(20,4))
    
    plt.subplot(1,3,1)
    cmap=sns.light_palette("blue")
    sns.heatmap(C,annot=True,fmt="d",cmap= cmap,xticklabels=[0,1],yticklabels=[0,1])
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Confusion matrix")
    
    plt.subplot(1,3,2)
    sns.heatmap(A,annot=True,fmt="f",cmap= cmap,xticklabels=[0,1],yticklabels=[0,1])
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Recall matrix")
    
    plt.subplot(1,3,3)
    sns.heatmap(B,annot=True,fmt="f",cmap= cmap,xticklabels=[0,1],yticklabels=[0,1])
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Precsion matrix")
    
    plt.show()
```

- Reference :

https://medium.com/analytics-vidhya/comparison-of-hyperparameter-tuning-algorithms-grid-search-random-search-bayesian-optimization-5326aaef1bd1

https://towardsdatascience.com/optimizing-hyperparameters-in-random-forest-classification-ec7741f9d3f6

- We have seen that for Tree Based Models which have several parameters to tune and number of estimators(number of individual trees) are too high,So using doing GridSearchCV or RandomizedSearchCV could  be too expesive on such a giant dataset for finding best parameters.In the previous method, we performed individual experiments by building multiple models with various hyperparameter values. All these experiments were independent of each other. Since each experiment was performed independently, we are not able to use the information from one experiment to improve the next experiment.


- Hence  we will be using the Bayesian Optimization Technique to tune the hyperparameters. Bayesian optimization is a sequential model-based optimization (SMO) algorithm that uses the results from the previous iteration to decide the next hyperparameter value candidates.


- So instead of blindly searching the hyperparameter space (like in grid search and random search), this method advocates the usage of intelligence to pick the next set of hyperparameters which will improve the model performance. We iteratively repeat this process until we converge to an optimum.



- An interesting analogy is to compare this to Bagging Vs Boosting. If you think about it, the idea is very similar!In bagging, we build a lot of trees in parallel and independent to each other. Boosting, on the other hand, is a sequential process where with each additional tree, we learn to correct the mistakes from its predecessor tree.



- Bayesian optimization creates a probabilistic model, mapping hyperparameters to a probability of a score on the objective function.

```python
#Code is inspired from 

#https://www.analyticsvidhya.com/blog/2021/05/bayesian-optimization-bayes_opt-or-hyperopt/
#https://towardsdatascience.com/kagglers-guide-to-lightgbm-hyperparameter-tuning-with-optuna-in-2021-ed048d9838b5
```

```python
from bayes_opt import BayesianOptimization
```

```python
def lightgbm_tuning_Bayesian_opt(num_leaves, max_depth,subsample, colsample_bytree, reg_alpha, reg_lambda,learning_rate):
    """
    A function to evaluate the lightgbm for HyperparmeterTuning
    """

    params = {
        'objective' : 'binary',
        'boosting_type' : 'gbdt',
        'learning_rate' : learning_rate,
        'num_leaves' : int(round(num_leaves)),
        'max_depth' : int(round(max_depth)),
        'subsample': subsample,
        'subsample_freq' : 1,
        'colsample_bytree' : colsample_bytree,
        'reg_alpha' : reg_alpha,
        'reg_lambda' : reg_lambda,
        'verbosity' : -1
    }
    
    
    
    
    lgbm_clf = LGBMClassifier(**params,n_estimators=10000,n_jobs = -1,random_state=42)

    fold = StratifiedKFold(n_splits = 3, shuffle = True, random_state = 33)
    
    cv_pred_proba= np.zeros(y_train.shape[0])
    
    for fold_k,(train_index,val_index) in enumerate(fold.split(X_train,y_train),1):
        
        #print("\t \t Starting K fold validation of fold number : ",fold_k)
        train_x=X_train.iloc[train_index]
        train_y=y_train.iloc[train_index]
        cv_x=X_train.iloc[val_index]
        cv_y=y_train.iloc[val_index]

            
        lgbm_clf.fit(train_x,train_y,eval_set=[(cv_x,cv_y)],early_stopping_rounds=100,eval_metric='auc',verbose=False)
        
        cv_pred_proba[val_index]= lgbm_clf.predict_proba(cv_x,num_iteration=lgbm_clf.best_iteration_)[:,1]
        
    return roc_auc_score(y_train,cv_pred_proba)
```

```python
params = {'colsample_bytree': (0.8, 1), #the fraction of columns to be randomly samples
          'learning_rate': (0.005, 0.01), 
          'num_leaves': (30, 40),  #max leaf nodes
          # 'n_estimators' :(100,10000),  #number of trees
          'subsample': (0.8, 1),  #fraction of observations to be randomly samples for each tree.
          'max_depth': (5, 10),  #maximum depth of a tree
          'reg_alpha': (.03, .05),  #L1 regularization
          'reg_lambda': (.06, .08), #L2 regularization
         }

bopt_lgbm = BayesianOptimization(lightgbm_tuning_Bayesian_opt, params,random_state=42)


bayesian_optimization = bopt_lgbm.maximize(n_iter = 6, init_points = 5)
```

```python
from random import choices
```

```python
def CustomStackingClassifier(X_train,y_train,X_cv,K,base_model,meta_model,p):
    
    #b) Now, in the 80% train set, split the train set into D1 and D2.(50-50)

    
    D1_X_train,D2_X_train,D1_y_train,D2_y_train=train_test_split(X_train,y_train,stratify=y_train,test_size=0.5,random_state=42)

    #From this D1, do sampling with replacement to create d1,d2,d3....dk(k samples).
    #SO for sampling we need to take sample from train and test both for training our 'K' models
    # but the problem is ,Anyhow we have less data for class 1, we have seen the class_weight is {0: 0.919, 1: 0.081}
    #So we will give weight as well while sampling,
    #The best way is if sampled index has class 0 we will give it very less weighteage i.e less sample size 
    # If its 1 we will give it more sample size
    #finally we can append it, It will work well as we are just sampling with replacement


    size_0= int(p*D1_y_train.value_counts()[0])
    size_1= int(D1_y_train.value_counts()[1])
    #Now here what problem i have faced is ,while we have sampled from 1 labeled data with more weights Hence The indexes wont be 
    #present in X_train data,SO better option is we drop y_train index and take index value from y_train which will be 0,1,..
    #for getting same index we can use index based location(.iloc) for getting corresponding X_train indexes 

    D1_y_train.reset_index(inplace=True,drop=True)

    k_sample_X=[]
    k_sample_y=[]

    for i in range(K):
        sample_indicies_0= random.choices(D1_y_train[D1_y_train==0].index.to_list(),k=size_0)
        sample_indicies_1= random.choices(D1_y_train[D1_y_train==1].index.to_list(),k=size_1)
        whole_Sample= sample_indicies_0+sample_indicies_1
        #whole_Sample= random.shuffle(whole_Sample)
        k_sample_X.append(D1_X_train.iloc[whole_Sample])
        k_sample_y.append(D1_y_train[whole_Sample])


    #Now create 'k' models and train each of these models with the samples. Each model will be trained on only one sample.
    #These models are called base models. We can select the best models and its hyperparameters based on the previous findings.
    #Make sure that ‘k’ can be passed as an argument.

    base_classifier=base_model

    k_trained_base_models=[]

    for i in range(K):
        base_classifier=base_model
        base_classifier.fit(k_sample_X[i],k_sample_y[i])
        k_trained_base_models.append(base_classifier)

    del k_sample_X,k_sample_y
    gc.collect()
    #c) Now pass the D2 set to each of these k models; now, you will get k predictions for D2 from each of these models.
    prediction_from_k_base_models=[]

    for i in range(K):
        pred_k= k_trained_base_models[i].predict_proba(D2_X_train)[:,1]
        prediction_from_k_base_models.append(pred_k)

    #d)1. Now, using these k predictions, create a new dataset, for D2, you already know it's corresponding target values,

    # prediction_from_k_base_models we have prediction as "row" from each K_th pretrained base model,
    #Because we need to create a dataset as each row as a feature or columns ,we can simply transpose from  rows to column

    new_datset_from_pred= np.array(prediction_from_k_base_models).transpose()

    # d) 2.so now you train a metamodel with these k predictions.

    metamodel_classifier=meta_model

    #fitting meta model classifier on new_datset_from_pred
    metamodel_classifier.fit(new_datset_from_pred,D2_y_train)

    #e.1. Now, for model evaluation, you can use the 20% data that you have kept as the test set. 
    #Pass that test set to each of the base models, and you will get 'k' predictions. 

    prediction_from_k_base_models_eval=[]

    for i in range(K):
        pred_eval_k=k_trained_base_models[i].predict_proba(X_cv)[:,1]
        prediction_from_k_base_models_eval.append(pred_eval_k)

    #e.2.Now you create a new dataset with these k predictions and pass it to your metamodel, and you will get the final prediction.
    new_datset_from_test= np.array(prediction_from_k_base_models_eval).transpose()

    final_test_prediction= metamodel_classifier.predict_proba(new_datset_from_test)[:,1]
    del new_datset_from_test
    gc.collect()
    #e.3 Using this final prediction and the targets for the test set, you can calculate the model's performance score.
    return final_test_prediction
```
